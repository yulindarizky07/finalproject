## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana hubungan antara musim dan jumlah sewa sepeda harian ?

```python
seasonal_averages = bike_sharing_data.groupby('season_daily')['cnt_daily'].mean()
season_names = ['Spring', 'Summer', 'Fall', 'Winter']
plt.bar(season_names, seasonal_averages)
plt.xlabel('Season')
plt.ylabel('Average Daily Bike Rentals')
plt.title('Impact of Season on Average Daily Bike Rentals')
plt.show()
```

### Pertanyaan 2: Apakah terdapat pola terkait waktu, misalnya berdasarkan bulan atau jam, dalam frekuensi penyewaan sepeda setiap hari?

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x="mnth_daily", y="cnt_daily", data=bike_sharing_data, ci=None)
plt.title("Pattern of Daily Bike Rentals Based on Month")
plt.xlabel("Month")
plt.ylabel("Daily Bike Rentals")
plt.show()
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x="hr", y="cnt_hourly", data=bike_sharing_data, ci=None)
plt.title("Pattern of Hourly Bike Rentals Based on Hour")
plt.xlabel("Hour")
plt.ylabel("Hourly Bike Rentals")
plt.show()
```
